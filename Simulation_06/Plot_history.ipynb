## 3 - Finding the bound

We know that for $d=1$, the cubic nonlinear Klein-Gordon equation with a defocusing $+u^3$ term and a standard (positive-mass) $-u$ term
\begin{aligned}
\begin{cases}
    \partial_{tt}u(z,t)-(i)^2\Delta u(z,t) = -u(z,t) + (u(z,t))^3 \qquad (z,t) \in \mathbb{C}^2 \times (0,\infty) \\
    u(z,0) = \phi(z) = \tanh(\frac{i}{2\sqrt{3}}(z_1+z_2))\\
    \partial_t u(z,0) = \psi(z) = -\sqrt{\frac{2}{3}}sech^2(\frac{i}{2\sqrt{3}}(z_1+z_2))
\end{cases}
\end{aligned}

admits a classical solution on the interval $[0,t_0]$ if both sufficient conditions are satisfied:

1. $\phi, \psi \in C^2(\overline{B}_{i}(z_0,t_0))$
2. $e^{3t_0^2} -e^{t_0^2} <  \Big(2 \max (1  , \Vert \phi \Vert_{\infty, \overline{B}_{i}(z_0,t_0)} + t_0 \Vert \psi \Vert_{\infty, \overline{B}_{i}(z_0,t_0)} + t_0 \Vert \nabla\phi \Vert_{\infty, \overline{B}_{i}(z_0,t_0)} )^2
\Big)^{-1}$

where 
$$
\overline{B}_{i}(z_0,t_0) :=\Bigl\{z\in\mathbb{C}^2:\exists y\in\mathbb{R}^2\text{ such that }z=z_0+iy \text{ and } |y|\le t_0 \Bigr\}
$$
given fixed $(z_0, t_0) \in \mathbb{C}^2 \times (0, \infty)$. 

Since our evaluation point is set at $(-1+0i, +1+0i)$, it is natural that we take $z_0=(-1+0i, -1+0i)$. Condition 1 is satisfied so long as the set $\overline{B}_{i}((-1,-1),t_0)$ does not contain a pole. Our objective is to find the maximum $t_0$ such that Condition 2 is also satisfied.

** Note: $z_0$ refers to a fixed 2-dimensional complex point in $\mathbb{C}^2$. Given a particular $z \in \mathbb{C}^2$, $z_1, z_2$ refers to its first and second complex coordinate respectively, i.e. $z=(z_1,z_2)$. So there's no confusion. **

For any $z\in \overline{B}_{i}((-1,-1),t_0)$, $z=(z_1,z_2)$ can be written as $z_1 = -1 + t_0 i y_1$ and $z_2 = -1 + t_0 i y_2$ where $(y_1, y_2) \in \mathbb{R}^2\cap B(0,1)$.

Now set $a = -\frac{t_0}{2\sqrt{3}}(y_1+y_2), b = -\frac{1}{\sqrt{3}}$ (both $a,b \in \mathbb{R}$), we have that 
$$
\phi(z)=\tanh(\frac{i}{2\sqrt{3}}(z_1+z_2))=\tanh(a+ib)
$$

Expanding function $\tanh$, we obtain
$$
\tanh(a+ib) = \frac{\sinh(2a)+i\sin(2b)}{\cosh(2a)+\cos(2b)}
$$
and for the modulus we have
$$
|\tanh(a+ib)|^2 = \frac{\sinh^2(2a)+\sin^2(2b)}{(\cosh(2a)+\cos(2b))^2}
$$
applying $\sinh(A) = \cosh(A)-1$ and $\sin^2(A)=1-\cos^2(A)$, we have
$$
|\tanh(a+ib)|^2 = \frac{\cosh^2(2a)-\cos^2(2b)}{(\cosh(2a)+\cos(2b))^2}=\frac{\cosh(2a)-\cos(2b)}{\cosh(2a)+\cos(2b)}
$$
Thus
$$
|\tanh(a+ib)| = \sqrt{\frac{\cosh(2a)-\cos(2b)}{\cosh(2a)+\cos(2b)}}
$$
Now, with $c=\cos(2b)=\cos(2/\sqrt{3})>0$ fixed, the function
$$
g(x)=\frac{x-c}{x+c} \qquad (x>c)
$$
is strictly increasing in $x$ since $g'(x) = \frac{2c}{(x+c)^2}>0$ (here c>0). Because $x=\cosh(2a)$ is even in $a$ and increases with $|a|$, the maximum of $|\tanh(a+ib)|$ is attained on the endpoints $a=\pm\frac{t_0}{2\sqrt{3}}\cdot \frac{2}{\sqrt{2}} = \pm\frac{t_0}{\sqrt{6}}$ $\Big($ i.e. $(y_1, y_2) = \pm(\frac{1}{\sqrt{2}},\frac{1}{\sqrt{2}}) \Big)$.
Putting together:
$$
\Vert \phi \Vert_{\infty, \overline{B}_{i}((-1,-1),t_0)} = \sqrt{\frac{\cosh(\frac{t_0}{\sqrt{6}}) - \cos(\frac{2}{\sqrt{3}})}{\cosh(\frac{t_0}{\sqrt{6}}) + \cos(\frac{2}{\sqrt{3}})}}
$$
which is a function of $t_0$.

Similarly, 
$$
\psi(z)=-\sqrt{\frac{2}{3}}sech^2(\frac{i}{2\sqrt{3}}(z_1+z_2))=-\sqrt{\frac{2}{3}}sech^2(a+ib)
$$
We can make use of the identity
$$
|sech^2(a+ib)|=\frac{2}{\cosh(2a)+\cos(2b)}
$$
where $\cosh$ attains its minimum at argument $0$. Thus, 
$$
\Vert \psi \Vert_{\infty, \overline{B}_{i}((-1,-1),t_0)} = \sqrt{\frac{2}{3}}\cdot\frac{2}{1+\cos(\frac{2}{\sqrt{3}})}=\frac{2\sqrt{\frac{2}{3}}}{\cos(\frac{2}{\sqrt{3}})+1} \approx 1.162941125285719
$$
and $\Vert \psi \Vert_{\infty, \overline{B}_{i}((-1,-1),t_0)}$ does not depend on $t_0$. 

Lastly, we know that 
$$
\nabla\phi(z) = \begin{bmatrix} \partial_{z_1}\tanh(\frac{i}{2\sqrt{3}}(z_1+z_2)) \\ \partial_{z_2}\tanh(\frac{i}{2\sqrt{3}}(z_1+z_2)) \end{bmatrix} 
= \begin{bmatrix} \frac{i}{2\sqrt{3}}sech^2(\frac{i}{2\sqrt{3}}(z_1+z_2)) \\ \frac{i}{2\sqrt{3}}sech^2(\frac{i}{2\sqrt{3}}(z_1+z_2)) \end{bmatrix}
= \frac{i}{2\sqrt{3}} \begin{bmatrix} sech^2(a+ib) \\ sech^2(a+ib) \end{bmatrix}
$$
Therefore, 
$$
\Vert \nabla\phi \Vert_{\infty, \overline{B}_{i}((-1,-1),t_0)} = \min_{a} \frac{1}{2\sqrt{3}} \sqrt{|sech^2(a+ib)|^2+|sech^2(a+ib)|^2} = \min_{a} \frac{1}{\sqrt{6}} |sech^2(a+ib)| = \frac{\sqrt{\frac{2}{3}}}{\cos(\frac{2}{\sqrt{3}})+1} \approx 0.5814705626428595
$$
and $\Vert \nabla\phi \Vert_{\infty, \overline{B}_{i}((-1,-1),t_0)}$ does not depend on $t_0$. 

Thus, overall our objective is to find the maximum $t_0$ such that 

$e^{3t_0^2} -e^{t_0^2} <  \Biggl(2 \max\Big(1, \sqrt{\frac{\cosh(\frac{t_0}{\sqrt{6}}) - \cos(\frac{2}{\sqrt{3}})}{\cosh(\frac{t_0}{\sqrt{6}}) + \cos(\frac{2}{\sqrt{3}})}} + t_0 \frac{2\sqrt{\frac{2}{3}}}{\cos(\frac{2}{\sqrt{3}})+1} + t_0 \frac{\sqrt{\frac{2}{3}}}{\cos(\frac{2}{\sqrt{3}})+1} \Big)^2
\Biggl)^{-1}$

```python
import math
import numpy as np

psi_bound = (2*math.sqrt(2/3)) / (math.cos(2/math.sqrt(3)) + 1)
print("L_infty bound for psi: ", psi_bound)

nabla_phi_bound = (math.sqrt(2/3)) / (math.cos(2/math.sqrt(3)) + 1)
print("L_infty bound for nabla phi: ", nabla_phi_bound)

def phi_bound(t):
    return math.sqrt((math.cosh(t/math.sqrt(6)) - math.cos(2/math.sqrt(3))) / (math.cosh(t/math.sqrt(6)) + math.cos(2/math.sqrt(3))))

def f(t):
    # t is a float scalar; phi_bound and psi_bound are expected to be defined earlier
    return math.exp(3*t**2) - math.exp(t**2) - 1.0/(2*max(1,(phi_bound(t) + t*psi_bound + t*nabla_phi_bound))**2)

def generate_T_vector(start=0.0, stop=0.2, steps=10001):
    """Generate a 1D array of t values from start to stop (inclusive) with 'steps' samples."""
    return np.linspace(start, stop, num=steps)

# --- User-adjustable parameters ---
start = 0.349
stop = 0.350
steps = 10001  # e.g. 10001 points gives step size 2e-5 over [0,0.2]

# Build T_vector and evaluate f(t)
T_vector = generate_T_vector(start, stop, steps)
i_values = np.array([f(float(t)) for t in T_vector])

# Find t values where f(t) < 0 and pick the maximum t (largest t that still satisfies the inequality)
valid_mask = i_values < 0
if not np.any(valid_mask):
    print('No t in the chosen range satisfies f(t) < 0')
    max_t_value = None
else:
    max_t_value = float(T_vector[valid_mask].max())
    print('Maximum t_0 value found:', max_t_value)

# Plot i_values against T_vector with a marker for the found maximum t0 (if any)
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.plot(T_vector, i_values, label='f(t) vs t', lw=1)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Plot of f(t) against t')
plt.axhline(0, color='k', linestyle='--')
if max_t_value is not None:
    plt.axvline(max_t_value, color='red', linestyle='--', label=f'max t0 = {max_t_value:.6f}')
plt.legend()
plt.grid(True)
plt.show()
```
